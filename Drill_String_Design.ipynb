## Drill String Designing




We are trying to check the design and comment on the safety of each section under normal design condition and normal design factor.

```python
# Taking input from the user for all variables
L = int(input("Enter the well depth in feet: "))
md = int(input('Enter the mud density in ppg: '))  # Mud density in ppg
MOP = int(input('Enter the Margin of Overpull in lbs: '))
WOB = int(input('Enter the Weight on Bit in lbs: '))
Pf = int(input('Enter the Formation Pressure in psi: '))
SFT = float(input('Enter the Safety Factor: '))

# Drill collar details
dec = float(input('Enter the drill collar outside diameter in inches: '))
dic = float(input('Enter the drill collar inside diameter in inches: '))
Lc = int(input('Enter the drill collar height in feet: '))

# Drill pipe details
dep = float(input('Enter the drill pipe outside diameter in inches: '))
dip = float(input('Enter the drill pipe inside diameter in inches: '))
wp = float(input('Enter the drill pipe weight in lbs/ft: '))

# Displaying the inputs
print("\nUser Inputs:")
print(f"Well Depth (L): {L} feet")
print(f"Mud Density (md): {md} ppg")
print(f"Margin of Overpull (MOP): {MOP} lbs")
print(f"Weight on Bit (WOB): {WOB} lbs")
print(f"Formation Pressure (Pf): {Pf} psi")
print(f"Safety Factor (SFT): {SFT}")
print(f"Drill Collar Outside Diameter (dec): {dec} inches")
print(f"Drill Collar Inside Diameter (dic): {dic} inches")
print(f"Drill Collar Height (Lc): {Lc} feet")
print(f"Drill Pipe Outside Diameter (dep): {dep} inches")
print(f"Drill Pipe Inside Diameter (dip): {dip} inches")
print(f"Drill Pipe Weight (wp): {wp} lbs/ft")
```

```python
t=(dec-dic)/2
print(t)
```

```python
wdc=10.68*t*(dec-t)
print(wdc)
```

```python
b=(65.4-md)/65.4
print(b)
```

```python
import pandas as pd



# Calculate the total length available for drill pipes
remaining_length = L - Lc
print(f"\nTotal length available for drill pipes: {remaining_length} feet")

# Taking the number of sections
num_sections = int(input("Enter the number of sections: "))

# Initialize lists to store user inputs
sections = []
grades = []
classes = []
weights = []
lengths = []
collapse_pressures = []

# Loop to gather inputs for each section
for i in range(num_sections):
    print(f"\nSection {i+1}:")

    section = int(input(f"Enter the section number: "))
    grade = input(f"Enter the grade: ")
    ym = int(input(f"Enter the yield strength (Ym): "))
    class_type = input(f"Enter the class: ")
    weight = float(input(f"Enter the weight (ppf): "))

    # Input the length and validate to ensure it doesn't exceed the remaining length
    length = int(input(f"Enter the length (ft) (Remaining: {remaining_length} ft): "))
    while length > remaining_length:
        print(f"Length exceeds remaining length. You have {remaining_length} feet remaining.")
        length = int(input(f"Re-enter the length (ft): "))

    collapse_pressure = int(input(f"Enter the collapse pressure, Pc (psi): "))

    # Append inputs to the respective lists
    sections.append(section)
    grades.append(grade)
    yms.append(ym)
    classes.append(class_type)
    weights.append(weight)
    lengths.append(length)
    collapse_pressures.append(collapse_pressure)

    # Subtract the entered length from the remaining length
    remaining_length -= length

    # Check if the remaining length is zero
    if remaining_length == 0:
        print("All available length has been allocated to the sections.")
        break

# Creating the dictionary with the inputs
data = {
    "Section": sections,
    "Grade": grades,
    "Ym": yms,
    "Class": classes,
    "Weight (ppf)": weights,
    "Length (ft)": lengths,
    "Collapse Pressure,Pc (psi)": collapse_pressures
}

# Creating DataFrame
df = pd.DataFrame(data)






df['Wa'] = (wdc*Lc+(df['Weight (ppf)'] * df['Length (ft)']).cumsum())*b# Load up to top of section lbs
# Display the DataFrame
df
```

```python
df.dtypes
```

```python
import pandas as pd

# Function to calculate values based on t and d
def calculate_values(t, d):
    # Class 1 (New)
    tst_new = t
    tsd_new = d
    pst_new = t
    psd_new= d

    # Class P (Premium)
    tst_premium = t * 0.8
    tsd_premium = d - t* 0.4
    pst_premium = t * 0.8
    psd_premium = d - t* 0.4
     # Class 2
    tst_2 =0.8*t
    tsd_2 = d-0.4*t
    pst_2 =0.65*t
    psd_2 = d-0.7*t
    # Class 3
    tst_3 = t * 0.625
    tsd_3 = d-0.75*t
    pst_3 = t * 0.55
    psd_3 = d-0.9*t

    # Additional classes can be added here following the same pattern

    # Create a dictionary with the calculated data
    data = {
        "Class": ["1","P","2","3"],
        "t for Tensile strength": [tst_new, tst_premium,tst_2,tst_3],
        "d for Tensile Strength": [tsd_new, tsd_premium,tsd_2,tsd_3],
        "t for Pressure": [pst_new, pst_premium,pst_2,pst_3],
        "d for Pressure": [psd_new, psd_premium,psd_2,psd_3],
    }

    # Create a DataFrame
    dt = pd.DataFrame(data)
    return dt

# Input values from user
t = float(input("Enter the value of t: "))
d = float(input("Enter the value of d: "))


# Calculate the table data
dt = calculate_values(t, d)

# Display the DataFrame
dt
```

```python
dt.dtypes
```

```python
import math

merged_df = pd.merge(df, dt, on="Class", how="left")

# Function to calculate pressure based on the merged data
def calculate_pressure(row):
    if row["Class"] == "1":
        return (1.75 *  row["t for Pressure"]  * row["Ym"]) / d
    else:
        return (2 *  row["t for Pressure"]  * row["Ym"]) / d



merged_df["Burst Pressure,Pb(psi)"] = merged_df.apply(calculate_pressure, axis=1)
merged_df['Tensile Strength,Pt(lbs)'] = merged_df['Ym'] * math.pi * merged_df['t for Tensile strength'] * (merged_df['d for Tensile Strength'] - merged_df['t for Tensile strength'])
merged_df['Pa'] = merged_df['Tensile Strength,Pt(lbs)']*0.9


selected_columns = ['Section', 'Grade', 'Class', 'Weight (ppf)','Collapse Pressure,Pc (psi)','Burst Pressure,Pb(psi)','Tensile Strength,Pt(lbs)','Wa','Pa']




filtered_df = merged_df.loc[:, selected_columns]



filtered_df
```

CROSS CHECKING FOR DRILL COLLAR

changing data type of columns burst pressure and tensile strength


```python
filtered_df['Burst Pressure,Pb(psi)']  = filtered_df['Burst Pressure,Pb(psi)'] .round(decimals=0)
filtered_df['Tensile Strength,Pt(lbs)']  = filtered_df['Tensile Strength,Pt(lbs)'] .round(decimals=0)
filtered_df['Wa']  = filtered_df['Wa'] .round(decimals=0)
filtered_df['Pa']  = filtered_df['Pa'] .round(decimals=0)

filtered_df['Burst Pressure,Pb(psi)'] = filtered_df['Burst Pressure,Pb(psi)'].astype(int)
filtered_df['Tensile Strength,Pt(lbs)'] = filtered_df['Tensile Strength,Pt(lbs)'].astype(int)
filtered_df['Wa']  = filtered_df['Wa'] .astype(int)
filtered_df['Pa']  = filtered_df['Pa'] .astype(int)
```

```python
filtered_df
```

```python
AWOB=int(Lc*(wdc-(1-b)*(wp-wdc))-L*wp*(1-b))
print(AWOB)

```

```python
if AWOB<WOB:
    print("Drill collar is unsafe")
else:
  print('Drill Collar is safe')
```

```python
filtered_df['AMOP'] = filtered_df['Pa'] - filtered_df['Wa']
filtered_df['ASFT'] = filtered_df['Pa']/filtered_df['Wa']
filtered_df

```

```python

filtered_df['Hydrostatic Pressure,Ph']= 0.052 * md * (L-Lc-df['Length (ft)'].cumsum().shift(1).fillna(0))


filtered_df
```

```python
filtered_df['ANc']=filtered_df['Collapse Pressure,Pc (psi)']/filtered_df['Hydrostatic Pressure,Ph']
filtered_df['ANi']=filtered_df['Burst Pressure,Pb(psi)']/Pf
filtered_df
```

```python
filtered_df['ANc'] = filtered_df['ANc'].round(decimals=3)
filtered_df['ANi'] = filtered_df['ANi'].round(decimals=3)

filtered_df['ASFT'] = filtered_df['ASFT'].round(decimals=2)
filtered_df
```

```python




# Define thresholds for each column
thresholds = {'AMOP': MOP, 'ASFT': SFT,'ANc':1.125,'ANi':1.000,'AWOB': WOB}

highlight_columns = set(thresholds.keys())

# Define a function to highlight cells based on column-specific thresholds
def highlight_cells(s):
    if s.name in highlight_columns:
        threshold = thresholds.get(s.name, float('inf'))  # Get threshold for the column
        return ['background-color: red' if isinstance(v, (int, float)) and v < threshold else '' for v in s]
    else:
        return ['' for _ in s]  # No style for other columns

# Apply the styling function column-wise
special_columns = ['Section', 'Grade', 'Class', 'AMOP','ASFT','ANc','ANi']




results_df= filtered_df.loc[:, special_columns]

# Display the styled DataFrame in Jupyter Notebook or JupyterLab



```
